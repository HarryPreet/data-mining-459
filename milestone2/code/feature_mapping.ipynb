```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
train= pd.read_csv('data/cases_2021_train_processed_2.csv')
test=pd.read_csv('data/cases_2021_test_processed_unlabelled_2.csv')
```

```python
print(np.sort(test['province'].unique()))
print(np.sort(train['province'].unique()))
```

```python
train['date_confirmation']=pd.to_datetime(train['date_confirmation'], format='%Y-%m-%d')
```

```python
train['date_confirmation']=train['date_confirmation'].astype(int)
```

```python
enc=OrdinalEncoder()
train['sex']=enc.fit_transform(train[['sex']])
test['sex']=enc.fit_transform(test[['sex']])
countries_train=enc.fit_transform(train[['country']])
train['outcome_group']=enc.fit_transform(train[['outcome_group']])
train['chronic_disease_binary']=enc.fit_transform(train[['chronic_disease_binary']])
```
